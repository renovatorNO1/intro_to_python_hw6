The bell curve in two dimensions has an analog of standard deviation called **covariance**

```python
import numpy as np
m1 = [2.5, 3.5]
m2 = [0.5, 1]
cov1 = [[1,1],[1,4.5]]
cov2 = [[2,0],[0,1]]
```

```python
x1 = np.random.multivariate_normal(m1, cov1, 200)
x1
```

```python
x2 = np.random.multivariate_normal(m2, cov2, 300)
x2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x1[:,0],x1[:,1], 'ro')
plt.show()
```

```python
plt.plot(x2[:,0], x2[:,1], 'bo')
plt.show()
```

```python
plt.plot(x1[:,0],x1[:,1],'ro',x2[:,0],x2[:,1],'bo')
plt.show()
```

Example of classifying situation:

* 600 observations in 2 classes, so 600 x 3 matrix (first column is class 1, second column is class 2, and third column is label)
  * 300 of the observations have one label and the other 300 have the other label
  * The training data will be 4/5 of the observations and their labels, so 480 x 3 matrix
    * Randomly shuffle the 600 x 3 matrix and partition into 5 120 x 3 matrices. On average, the labels should be evenly distributed
* Classifier(training, test) --> should return a 120 x 1 matrix of labels
  * Compare this 120 x 1 matrix to the labels of the last 120 observations in the 600 x 3 matrix

**Applying method to our data**

```python
x1.shape
```

```python
x2.shape
```

```python
# give x1 label "1"
labels1 = np.ones(200).reshape(200,1)
x1_labels = np.hstack((x1, labels1))
x1_labels.shape
```

```python
# give x2 label "2"
labels2 = 2 * np.ones(300).reshape(300,1)
x2_labels = np.hstack((x2, labels2))
x2_labels.shape
```

```python
# Now stack the two matrices with labels vertically to form a 500 x 3 matrix
data = np.vstack((x1_labels, x2_labels))
data
```

```python
# Now randomly shuffle the data
# Notice that np.random.shuffle() has a side effect
np.random.shuffle(data)
data

# We now have our data (the large matrix of all observations and their labels)
```

The following demonstrates how to use delete function

```python
a = np.arange(16).reshape(4,4)
a
```

Axis in numpy arrays: axis=0 is rows and axis=1 is columns

```python
# Killing row 2 of the matrix
# Second argument indicates row/column numbers, third argument is the axis number
np.delete(a, 1, 0)
```

```python
# Notice that this does not have a side effect
a
```

```python
# Killing second and third rows
# Notice that multiple elements as an argument must be in tuple form
np.delete(a, (1,2), 0)
```

```python
# Note that axis specification is optional. Not specifying axis flattens the matrix and the
# element number/numbers indicate element number/numbers in that 1 dimensional matrix
np.delete(a, (1,2))
```

As it pertains to the classifying example from earlier:

* Kill the first 400 rows and then the third column of the result to generate test matrix
